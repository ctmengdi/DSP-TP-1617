```python
%matplotlib inline
%pprint off

# To make some nice plot
import matplotlib.pyplot as plt
# To play with arrays
import numpy as np
import scipy as sp
from skimage import io

from skimage import img_as_float
from matplotlib import cm
from skimage import img_as_ubyte
from scipy import signal

```

## Filtering 
**1.1**

```python
b1, a1 = signal.butter(3, 100, btype='lowpass', analog=True)
w1, h1 = signal.freqs(b1, a1)

b2, a2 = signal.butter(3, 100, btype='highpass', analog=True)
w2, h2 = signal.freqs(b2, a2)

b3, a3 = signal.butter(3, [100, 150], btype='bandpass', analog=True)
w3, h3 = signal.freqs(b3, a3)

b4, a4 = signal.butter(3, [100, 150], btype='bandstop', analog=True)
w4, h4 = signal.freqs(b4, a4)

f, axarr = plt.subplots(2, 2)
axarr[0, 0].semilogx(w1, 20 * np.log10(abs(h1)))
axarr[0, 0].set_title('lowpass')
axarr[0, 0].margins(0, 0.1)
#axarr[0, 0].xscale('log')
axarr[0, 1].semilogx(w2, 20 * np.log10(abs(h2)))
axarr[0, 1].set_title('highpass')
axarr[0, 1].margins(0, 0.1)
#axarr[0, 1].xscale('log')
axarr[1, 0].semilogx(w3, 20 * np.log10(abs(h3)))
axarr[1, 0].set_title('bandpass')
axarr[1, 0].margins(0, 0.1)
#axarr[1, 0].xscale('log')
axarr[1, 1].semilogx(w4, 20 * np.log10(abs(h4)))
axarr[1, 1].set_title('bandstop')
axarr[1, 1].margins(0, 0.1)
#axarr[1, 1].xscale('log')

plt.tight_layout()
```

```python
b1, a1 = signal.cheby2(3, 5, 0.6, btype = 'lowpass')
w1, h1 = signal.freqz(b1, a1)

b2, a2 = signal.cheby2(5, 5, 0.6, btype = 'lowpass')
w2, h2 = signal.freqz(b2, a2)

b3, a3 = signal.cheby2(10, 5, 0.6, btype = 'lowpass')
w3, h3 = signal.freqz(b3, a3)

b4, a4 = signal.cheby2(20, 5, 0.6, btype = 'lowpass')
w4, h4 = signal.freqz(b4, a4)


f, axarr = plt.subplots(2, 2)
axarr[0, 0].plot(w1/np.pi, np.abs(h1))
axarr[0, 0].set_title('3rd lowpass')
axarr[0, 0].margins(0, 0.1)
axarr[0, 1].plot(w2/np.pi, np.abs(h2))
axarr[0, 1].set_title('5th lowpass')
axarr[0, 1].margins(0, 0.1)
axarr[1, 0].plot(w3/np.pi, np.abs(h3))
axarr[1, 0].set_title('10th lowpass')
axarr[1, 0].margins(0, 0.1)
axarr[1, 1].plot(w4/np.pi, np.abs(h4))
axarr[1, 1].set_title('20th lowpass')
axarr[1, 1].margins(0, 0.1)

plt.tight_layout()
```
